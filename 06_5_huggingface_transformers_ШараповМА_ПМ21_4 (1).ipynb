````python
context = """The seminars on Deep Learning and Natural Language Processing were truly captivating,
providing a deep dive into the intricacies of these disciplines.
The wealth of knowledge and insights gained during the sessions was commendable.
However, it's disheartening to note the scarcity of homework assignments.
Anastasia, in particular, is quite concerned that the limited number of assignments might
fall short of even reaching 30. While the seminars were intellectually stimulating,
the desire for more hands-on practice through assignments remains strong,
as it is crucial for reinforcing the theoretical understanding acquired during the classes."""
````

````python
question1 = "What would be the ideal number of homework assignments for Anastasia"
question2 = "What are the shortcomings of the course?"
````

````python
from transformers import pipeline

question_answerer = pipeline(
    "question-answering",
    model='distilbert-base-cased-distilled-squad'
)

result = question_answerer(
    question=question1,
    context=context
)
result
````

–ü–æ–ª—É—á–∏—Ç–µ –ø—Ä–æ–≥–Ω–æ–∑ –¥–ª—è –≤—Å–µ—Ö —Ç–µ–∫—Å—Ç–æ–≤ –∏–∑ —Ñ–∞–π–ª–∞ –∏ –ø–æ—Å—á–∏—Ç–∞–π—Ç–µ F1-score. –î–ª—è —É—Å–∫–æ—Ä–µ–Ω–∏—è —Ä–∞–±–æ—Ç—ã –º–æ–¥–µ–ª–∏ –≤—ã –º–æ–∂–µ—Ç–µ –ø–µ—Ä–µ–Ω–µ—Å—Ç–∏ –µ–µ –Ω–∞ GPU –∏ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –≤–æ–∑–º–æ–∂–Ω–æ—Å—Ç–∏ pipeline –¥–ª—è —Ä–∞–±–æ—Ç—ã —Å –±–∞—Ç—á–∞–º–∏ –∏ –≥–µ–Ω–µ—Ä–∞—Ç–æ—Ä–∞–º–∏.

- [ ] –ü—Ä–æ–≤–µ—Ä–µ–Ω–æ –Ω–∞ —Å–µ–º–∏–Ω–∞—Ä–µ

````python
import pandas as pd
file_path = "reviews_polarity.csv"
df = pd.read_csv(file_path)
````

````python
from transformers import pipeline
#–°—Ä–µ–¥–∏ –ø—Ä–µ–¥–æ–±—É—á–µ–Ω–Ω—ã—Ö –º–æ–¥–µ–ª–µ–π –Ω–∞–π–¥–∏—Ç–µ –º–æ–¥–µ–ª—å –¥–ª—è –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏ —Ç–æ–Ω–∞–ª—å–Ω–æ—Å—Ç–∏ —Ä—É—Å—Å–∫–æ—è–∑—ã—á–Ω–æ–≥–æ —Ç–µ–∫—Å—Ç–∞
classifier = pipeline("sentiment-analysis", model="sismetanin/rubert-toxic-pikabu-2ch")
````

````python
#–ü—Ä–æ—Ç–µ—Å—Ç–∏—Ä—É–π—Ç–µ –¥–∞–Ω–Ω—É—é –º–æ–¥–µ–ª—å –Ω–∞ –Ω–µ—Å–∫–æ–ª—å–∫–∏—Ö –ø—Ä–µ–¥–ª–æ–∂–µ–Ω–∏—è—Ö, –∏—Å–ø–æ–ª—å–∑—É—è transformers.pipeline. –í—ã–≤–µ–¥–∏—Ç–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã —Ä–∞–±–æ—Ç—ã –≤ —Å–ª–µ–¥—É—é—â–µ–º –≤–∏–¥–µ:
sentences = [
    "–°–ø–∞—Å–∏–±–æ, –æ—á–µ–Ω—å –≤–∫—É—Å–Ω–æ",
    "–£–∂–∞—Å–Ω–æ, –Ω–µ –≤–µ—Ä–Ω—É—Å—å —Å—é–¥–∞ –Ω–∏–∫–æ–≥–¥–∞, –≤—ã —É—Ä–æ–¥—ã",
]

for sentence in sentences:
    result = classifier(sentence)[0]
    print(f"{sentence} -> {result['label']} ({result['score']:.4f})")
````

````python
import torch
from sklearn.metrics import f1_score
from sklearn.model_selection import train_test_split
````

````python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

classifier = pipeline(
    "sentiment-analysis",
    model='sismetanin/rubert-toxic-pikabu-2ch',
    device=device
)

df['class'] = df['class'].astype(str)
````

````python
y_true = df['class'].tolist()
y_pred = [result['label'] for result in classifier(df['text'].tolist())]
y_pred = [label.replace('LABEL_', '') for label in y_pred]
````

````python
f1 = f1_score(y_true, y_pred, average='weighted')

print(f"F1-score: {f1:.4f}")
````

2\. –°–æ–∑–¥–∞–π—Ç–µ —Ç–æ–∫–µ–Ω–∏–∑–∞—Ç–æ—Ä, —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–∏–π –º–æ–¥–µ–ª–∏ –∏–∑ –ø—Ä–µ–¥—ã–¥—É—â–µ–≥–æ –∑–∞–¥–∞–Ω–∏—è, –∏—Å–ø–æ–ª—å–∑—É—è –∫–ª–∞—Å—Å `AutoTokenizer`. –í–æ–∑—å–º–∏—Ç–µ –æ–¥–Ω–æ –ø—Ä–µ–¥–ª–æ–∂–µ–Ω–∏–µ –∏–∑ –Ω–∞–±–æ—Ä–∞ –¥–∞–Ω–Ω—ã—Ö –∏ —Ç–æ–∫–µ–Ω–∏–∑–∏—Ä—É–π—Ç–µ –µ–≥–æ, –∏—Å–ø–æ–ª—å–∑—É—è —Å–æ–∑–¥–∞–Ω–Ω—ã–π –æ–±—ä–µ–∫—Ç. –í—ã–≤–µ–¥–∏—Ç–µ –Ω–∞ —ç–∫—Ä–∞–Ω –ø–æ–ª—É—á–µ–Ω–Ω—ã–π —Ä–µ–∑—É–ª—å—Ç–∞—Ç. –í—ã–ø–æ–ª–Ω–∏—Ç–µ –æ–±—Ä–∞—Ç–Ω–æ–µ –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ: –ø–æ–ª—É—á–∏—Ç–µ —Ç–µ–∫—Å—Ç –ø–æ –Ω–∞–±–æ—Ä—É —Ç–æ–∫–µ–Ω–æ–≤ –∏ –≤—ã–≤–µ–¥–∏—Ç–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç –Ω–∞ —ç–∫—Ä–∞–Ω.

–ü–æ–ª—É—á–∏—Ç–µ –±–∞—Ç—á –∏–∑ 10 –ø—Ä–µ–¥–ª–æ–∂–µ–Ω–∏–π –∏ —Ç–æ–∫–µ–Ω–∏–∑–∏—Ä—É–π—Ç–µ –µ–≥–æ. –ü—Ä–æ–¥–µ–º–æ–Ω—Å—Ç—Ä–∏—Ä—É–π—Ç–µ –≤–æ–∑–º–æ–∂–Ω–æ—Å—Ç–∏ —Ç–æ–∫–µ–Ω–∏–∑–∞—Ç–æ—Ä–∞ –¥–ª—è –ø–∞–¥–¥–∏–Ω–≥–∞, –æ–±—Ä–µ–∑–∫–∏, –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏—è –≤ —Ç–µ–Ω–∑–æ—Ä—ã.
–†–µ—à–∏—Ç–µ –∑–∞–¥–∞—á—É 1, —Å–æ–∑–¥–∞–≤ –æ–±—ä–µ–∫—Ç —Ç–æ–∫–µ–Ω–∏–∑–∞—Ç–æ—Ä–∞ (`AutoTokenizer`) –∏ –º–æ–¥–µ–ª—å (`AutoModelForSequenceClassification`).

- [ ] –ü—Ä–æ–≤–µ—Ä–µ–Ω–æ –Ω–∞ —Å–µ–º–∏–Ω–∞—Ä–µ

````python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
````

````python
model_name = 'sismetanin/rubert-toxic-pikabu-2ch'
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)

# –ü—Ä–∏–º–µ—Ä —Ç–æ–∫–µ–Ω–∏–∑–∞—Ü–∏–∏ –∏ –æ–±—Ä–∞—Ç–Ω–æ–≥–æ –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏—è –æ–¥–Ω–æ–≥–æ –ø—Ä–µ–¥–ª–æ–∂–µ–Ω–∏—è
sentence = "–°–ø–∞—Å–∏–±–æ, –æ—á–µ–Ω—å –≤–∫—É—Å–Ω–æ"
tokens = tokenizer(sentence)
decoded_sentence = tokenizer.decode(tokens['input_ids'])
print(f"–ò—Å—Ö–æ–¥–Ω–æ–µ –ø—Ä–µ–¥–ª–æ–∂–µ–Ω–∏–µ: {sentence}")
print(f"–¢–æ–∫–µ–Ω—ã: {tokens}")
print(f"–û–±—Ä–∞—Ç–Ω–æ–µ –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ: {decoded_sentence}")
````

````python
batch_size = 10
sample_df = df.sample(n=batch_size, random_state=42)

batch_sentences = sample_df['text'].tolist()
batch_labels = sample_df['class'].tolist()

# –¢–æ–∫–µ–Ω–∏–∑–∞—Ü–∏—è
batch_encoding = tokenizer(batch_sentences, padding=True, truncation=True, return_tensors='pt')

print("–¢–æ–∫–µ–Ω—ã:")
print(batch_encoding)
````

````python
# –í—ã–≤–æ–¥ –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏ –æ —Ç–µ–Ω–∑–æ—Ä–∞—Ö
input_ids = batch_encoding['input_ids']
attention_mask = batch_encoding['attention_mask']
print("–¢–µ–Ω–∑–æ—Ä input_ids:", input_ids)
print("–¢–µ–Ω–∑–æ—Ä attention_mask:", attention_mask)
````

````python
sample_df['class'] = sample_df['class'].astype(str)
batch_sentences = sample_df['text'].tolist()
batch_labels = torch.tensor([int(label) for label in sample_df['class'].tolist()])

# –¢–æ–∫–µ–Ω–∏–∑–∞—Ü–∏—è
batch_encoding = tokenizer(batch_sentences, padding=True, truncation=True, return_tensors='pt')

# –ú–æ–¥–µ–ª—å
output = model(**batch_encoding, labels=batch_labels)

# –í—ã–≤–æ–¥ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞
print("–õ–æ–≥–∏—Ç—ã –º–æ–¥–µ–ª–∏ (–±–µ–∑ Softmax):", output.logits)
print("–ü—Ä–æ–≥–Ω–æ–∑—ã –º–æ–¥–µ–ª–∏ (—Å Softmax):", torch.nn.functional.softmax(output.logits, dim=1))
````

````python
class ClassificationModel(nn.Module):
    def __init__(self):
        super(ClassificationModel, self).__init__()
        self.model = AutoModelForSequenceClassification.from_pretrained("blanchefort/rubert-base-cased-sentiment")
        for param in self.model.base_model.parameters():
            param.requires_grad = False
        self.model.classifier = nn.Linear(in_features=self.model.classifier.in_features, out_features=3, bias=True)
        self.loss = nn.CrossEntropyLoss()

    def forward(self, input_ids, labels=None):
        outputs = self.model(input_ids, labels=labels)
        logits = outputs.logits
        loss = outputs.loss if labels is not None else None

        return SequenceClassifierOutput(
            loss=loss,
            logits=logits,
        )
````

````python
class Classifier(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super(Classifier, self).__init__()
        self.fc1 = nn.Linear(input_size, hidden_size)
        self.relu = nn.ReLU()
        self.fc2 = nn.Linear(hidden_size, output_size)

    def forward(self, x):
        x = self.fc1(x)
        x = self.relu(x)
        x = self.fc2(x)
        return x
````

````python
input_size = transformer.get_sentence_embedding_dimension()
hidden_size = 256
output_size = 2
````

````python
from datasets import load_dataset
data_files = {'train': 'train.csv', 'test': 'test.csv'}
ds = load_dataset('/content/reviews_polarity_dataset', data_files=data_files)
````

````python
ds = ds.remove_columns('Unnamed: 0')
````

````python
loader_train = DataLoader(ds['train'], batch_size=64)
loader_test = DataLoader(ds['test'], batch_size=64)
````

````python
classifier_model = Classifier(input_size, hidden_size, output_size)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
classifier_model = classifier_model.to(device)

optimizer = AdamW(classifier_model.parameters(), lr=5e-5)
n_epochs = 5
scheduler = get_linear_schedule_with_warmup(optimizer, num_warmup_steps=0, num_training_steps=n_epochs * len(loader_train))

for epoch in range(n_epochs):
    classifier_model.train()
    batch_loss = []

    for batch in DataLoader(ds['train'], batch_size=64, shuffle=True):
        inputs = transformer.encode(batch['text'], convert_to_tensor=True).to(device)
        labels = batch['class'].to(device)

        optimizer.zero_grad()
        outputs = classifier_model(inputs)

        loss = nn.CrossEntropyLoss()(outputs, labels)
        loss.backward()
        optimizer.step()
        scheduler.step()

        batch_loss.append(loss.item())

    print(f'Epoch: {epoch + 1}, Loss: {sum(batch_loss) / len(batch_loss)}')

    classifier_model.eval()
    predictions = []
    true_labels = []

    with torch.no_grad():
        for batch in DataLoader(ds['test'], batch_size=64, shuffle=False):
            inputs = transformer.encode(batch['text'], convert_to_tensor=True).to(device)
            labels = batch['class'].to(device)

            outputs = classifier_model(inputs)
            logits = outputs
            preds = torch.argmax(logits, dim=1)

            predictions.extend(preds.cpu().numpy())
            true_labels.extend(labels.cpu().numpy())

    f1_test = f1_score(true_labels, predictions, average='binary')
    print(f'F1 score on the test set: {f1_test}')
````
